##**DATA CLEANING**



```python
import pandas as pd

def clean_dataset(df, city_name):
    # Step 1: Standardize column names
    df.columns = [col.strip().upper() for col in df.columns]

    # Step 2: Column mapping (handles different datasets)
    column_mapping = {
        'PRICE_PER_UNIT_AREA': ['PRICE_PER_UNIT_AREA', 'PRICE', 'PRICE_PER_SQFT', 'PRICE_PER_UNIT'],
        'BEDROOM_NUM': ['BEDROOM_NUM', 'BHK', 'BEDROOMS'],
        'PROPERTY_TYPE': ['PROPERTY_TYPE', 'TYPE', 'PROPERTY']
    }

    cleaned_df = pd.DataFrame()

    for standard_col, possible_names in column_mapping.items():
        for col in possible_names:
            if col in df.columns:
                cleaned_df[standard_col] = df[col]
                break

    # Step 3: Keep only required columns
    required_cols = ['PRICE_PER_UNIT_AREA', 'BEDROOM_NUM', 'PROPERTY_TYPE']
    cleaned_df = cleaned_df[[col for col in required_cols if col in cleaned_df.columns]]

    # Step 4: Handle missing values
    cleaned_df.dropna(inplace=True)

    # Step 5: Convert data types
    cleaned_df['PRICE_PER_UNIT_AREA'] = pd.to_numeric(cleaned_df['PRICE_PER_UNIT_AREA'], errors='coerce')
    cleaned_df['BEDROOM_NUM'] = pd.to_numeric(cleaned_df['BEDROOM_NUM'], errors='coerce')

    cleaned_df.dropna(inplace=True)

    # Step 6: Add city label
    cleaned_df['CITY'] = city_name

    return cleaned_df
```

```python
# Clean each dataset
mumbai_clean = clean_dataset(mumbai, "MUMBAI")
hyderabad_clean = clean_dataset(hyderabad, "HYDERABAD")
kolkata_clean = clean_dataset(kolkata, "KOLKATA")
gurgaon_clean = clean_dataset(gurgaon, "GURGAON")


# mumbia['city'] = 'mumbia'
# hyderabad['city'] = 'hyderabad'
# kolkata['city'] = 'kolkata'
# gurgaon['city'] = 'gurgaon'


print(mumbai_clean.head(1))
print(hyderabad_clean.head(1))
print(kolkata_clean.head(1))
print(gurgaon_clean.head(1))
```

##ONE WAY ANOVA

```python
def one_way_anova(data, group_col, value_col):
    groups = data[group_col].unique()
    grand_mean = data[value_col].mean()

    SSB, SSW = 0, 0
    N = len(data)
    k = len(groups)

    for g in groups:
        vals = data[data[group_col]==g][value_col]
        mean = vals.mean()
        n = len(vals)

        SSB += n * (mean - grand_mean)**2
        SSW += sum((vals - mean)**2)

    dfb = k - 1
    dfw = N - k

    MSB = SSB / dfb
    MSW = SSW / dfw

    F = MSB / MSW

    return F, dfb, dfw

F, d1, d2 = one_way_anova(final_data, 'CITY', 'PRICE_PER_UNIT_AREA')
print(F)
```

```python
import scipy.special # Import scipy for numerically stable beta function

def beta(a,b):
    # Use scipy's beta function to avoid OverflowError with large inputs
    return scipy.special.beta(a,b)

def f_pdf(x,d1,d2):
    return ((d1/d2)**(d1/2)*x**(d1/2-1)) / (beta(d1/2,d2/2)*(1+(d1/d2)*x)**((d1+d2)/2))

def p_value(F,d1,d2):
    step=1000
    width=F/step
    area=0
    for i in range(step):
        x=i*width
        area+=f_pdf(x,d1,d2)*width
    return 1-area

p = p_value(F,d1,d2)
print(p)
```

# 📊 Results Interpretation

- If p < 0.05 → Significant difference  
- If p ≥ 0.05 → No significant difference  

### Insight

If significant:

👉 Property prices vary across cities  
👉 Location plays a major role  

If not:

👉 Prices are similar across cities  

```python
if p < 0.05:
    print("✅ Significant difference: City affects price")
else:
    print("❌ No significant difference")
```

##Two-Way ANOVA

# 🧠 Two-Way ANOVA (From Scratch)

## 🎯 Objective

To analyze the effect of **two factors simultaneously**:

- Factor A → CITY  
- Factor B → PROPERTY TYPE  

on:

- Dependent Variable → PRICE_PER_UNIT_AREA  


## 🎯 Interpretation

- Large F → Significant effect  
- Small F → No significant effect  

```python
def two_way_anova_full(data, factor1, factor2, value):

    grand_mean = data[value].mean()

    levels1 = data[factor1].unique()   # CITY
    levels2 = data[factor2].unique()   # PROPERTY TYPE

    SSA, SSB, SSE = 0, 0, 0

    # Factor A (City)
    for a in levels1:
        subset = data[data[factor1] == a]
        mean_a = subset[value].mean()
        SSA += len(subset) * (mean_a - grand_mean) ** 2

    # Factor B (Property Type)
    for b in levels2:
        subset = data[data[factor2] == b]
        mean_b = subset[value].mean()
        SSB += len(subset) * (mean_b - grand_mean) ** 2

    # Error (interaction + residual)
    for a in levels1:
        for b in levels2:
            subset = data[(data[factor1] == a) & (data[factor2] == b)]
            if len(subset) > 0:
                mean_ab = subset[value].mean()
                SSE += sum((subset[value] - mean_ab) ** 2)

    # Degrees of freedom
    dfA = len(levels1) - 1
    dfB = len(levels2) - 1
    dfE = len(data) - (len(levels1) * len(levels2))

    # Mean squares
    MSA = SSA / dfA
    MSB = SSB / dfB
    MSE = SSE / dfE

    # F-statistics
    FA = MSA / MSE
    FB = MSB / MSE

    return {
        "SSA": SSA,
        "SSB": SSB,
        "SSE": SSE,
        "FA": FA,
        "FB": FB,
        "dfA": dfA,
        "dfB": dfB,
        "dfE": dfE
    }

# Run Two-Way ANOVA
result_2way = two_way_anova_full(final_data, 'CITY', 'PROPERTY_TYPE', 'PRICE_PER_UNIT_AREA')

print(result_2way)
```

```python
FA = result_2way["FA"]
FB = result_2way["FB"]

dfA = result_2way["dfA"]
dfB = result_2way["dfB"]
dfE = result_2way["dfE"]

pA = p_value(FA, dfA, dfE)
pB = p_value(FB, dfB, dfE)

print("City Effect p-value:", pA)
print("Property Type Effect p-value:", pB)
```

# 📊 Two-Way ANOVA Results Interpretation

## 🔹 City Effect
- If p < 0.05 → City significantly affects price  

## 🔹 Property Type Effect
- If p < 0.05 → Property type significantly affects price  

---

## 🎯 Insight

- If both significant → Multiple factors influence price  
- If only one significant → Dominant factor identified  
- If none → No strong effect  

This helps in **real estate decision-making**.

```python
alpha = 0.05

print("\n===== FINAL TWO-WAY ANOVA DECISION =====")

if pA < alpha:
    print("✅ City has a significant effect on price")
else:
    print("❌ City has NO significant effect")

if pB < alpha:
    print("✅ Property Type has a significant effect on price")
else:
    print("❌ Property Type has NO significant effect")
```

##MANOVA

# 🧠 MANOVA (Multivariate Analysis of Variance)

## 🎯 Objective

To test whether **multiple dependent variables jointly differ across groups**.

In this study:

- Independent Variable → CITY  
- Dependent Variables →  
  - PRICE_PER_UNIT_AREA  
  - BEDROOM_NUM  

---

## 🧮 Mathematical Idea

Instead of analyzing variables separately, MANOVA analyzes them **together**.

We compute:

- Within-group variation → **W matrix**
- Between-group variation → **B matrix**

---

## 📐 Wilks’ Lambda (Λ)

\[
\Lambda = \frac{|W|}{|T|}
\]

Where:

- \( W \) = Within-group covariance matrix  
- \( T = W + B \) = Total covariance matrix  

---

## 🎯 Interpretation

- Λ close to **0** → Strong difference between groups  
- Λ close to **1** → Groups are similar  

---

## 💡 Why MANOVA?

Because real-world variables (price, bedrooms) are **correlated**,  
so analyzing them together gives better insights.

```python
import numpy as np

def manova_full(df, group_col, dependent_vars):

    groups = df[group_col].unique()
    X = df[dependent_vars].values

    # Grand mean vector
    grand_mean = np.mean(X, axis=0)

    # Initialize matrices
    W = np.zeros((len(dependent_vars), len(dependent_vars)))
    B = np.zeros((len(dependent_vars), len(dependent_vars)))

    for g in groups:
        group_data = df[df[group_col] == g][dependent_vars].values
        n_g = group_data.shape[0]

        group_mean = np.mean(group_data, axis=0)

        # Within-group scatter matrix
        for row in group_data:
            diff = (row - group_mean).reshape(-1, 1)
            W += diff @ diff.T

        # Between-group scatter matrix
        mean_diff = (group_mean - grand_mean).reshape(-1, 1)
        B += n_g * (mean_diff @ mean_diff.T)

    # Total matrix
    T = W + B

    # Determinants
    det_W = np.linalg.det(W)
    det_T = np.linalg.det(T)

    # Wilks' Lambda
    wilks_lambda = det_W / det_T if det_T != 0 else 0

    return {
        "Wilks_Lambda": wilks_lambda,
        "W_matrix": W,
        "B_matrix": B,
        "T_matrix": T
    }

# Run MANOVA
result_manova = manova_full(final_data, 'CITY', ['PRICE_PER_UNIT_AREA', 'BEDROOM_NUM'])

print("Wilks' Lambda:", result_manova["Wilks_Lambda"])
```

# 📊 MANOVA Results Interpretation

## 🎯 Decision Rule

- If Λ < 0.05 → Strong multivariate difference  
- If 0.05 ≤ Λ < 0.3 → Moderate difference  
- If Λ ≥ 0.3 → Weak or no difference  

---

## 💡 Meaning

- Low Lambda → Cities differ across multiple features  
- High Lambda → Cities are similar  

---

## 🧠 Insight

This tells us whether **overall property characteristics vary across cities**,  
not just one variable at a time.

```python
lambda_val = result_manova["Wilks_Lambda"]

print("\n===== MANOVA FINAL RESULT =====")

if lambda_val < 0.05:
    print("✅ Strong multivariate effect")
    print("👉 Cities differ significantly across price and bedroom features")
elif lambda_val < 0.3:
    print("⚠️ Moderate effect")
    print("👉 Cities show noticeable differences")
else:
    print("❌ Weak effect")
    print("👉 Cities are similar across variables")
```
